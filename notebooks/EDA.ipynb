```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
data = pd.read_csv("../data/credit_data.csv")
print("Dataset Shape:", data.shape)
print("\nFirst 5 Rows:\n", data.head())
print("\nData Info:\n")
data.info()
```

```python
# Check for missing values
print("Missing Values:\n", data.isnull().sum())
```

```python
# Summary statistics for numerical features
print("Summary Statistics:\n", data.describe())
```

```python
# Plot distribution of defaults
plt.figure(figsize=(6, 4))
sns.countplot(x="default payment next month", data=data)
plt.title("Distribution of Loan Defaults (0 = No, 1 = Yes)")
plt.xlabel("Default Status")
plt.ylabel("Count")
plt.show()
```

```python
# Distribution of credit limit by default status
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="LIMIT_BAL", hue="default payment next month", bins=30, kde=True)
plt.title("Credit Limit Distribution by Default Status")
plt.xlabel("Credit Limit (LIMIT_BAL)")
plt.ylabel("Count")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=False, cmap="coolwarm")
plt.title("Feature Correlations")
plt.show()
```

```python
# Default rate by sex
plt.figure(figsize=(6, 4))
sns.countplot(x="SEX", hue="default payment next month", data=data)
plt.title("Default Counts by Sex (1 = Male, 2 = Female)")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
# Age distribution by default status
plt.figure(figsize=(10, 6))
sns.boxplot(x="default payment next month", y="AGE", data=data)
plt.title("Age Distribution by Default Status")
plt.xlabel("Default Status")
plt.ylabel("Age")
plt.show()
```

```python
# Cell: Check column names
print("Columns in dataset:\n", data.columns.tolist())
```

```python
# Boxplot of payment status vs default
plt.figure(figsize=(10, 6))
sns.boxplot(x="default payment next month", y="PAY_0", data=data)
plt.title("Payment Status (PAY_0) vs Default")
plt.xlabel("Default Status")
plt.ylabel("PAY_0 (Payment Delay)")
plt.show()
```

```python
# Load the dataset
data = pd.read_csv("../data/credit_data.csv")
print("Dataset Shape:", data.shape)
print("\nFirst 5 Rows:\n", data.head())
print("\nData Info:\n")
data.info()
```

```python
# Boxplot of payment status vs default
plt.figure(figsize=(10, 6))
sns.boxplot(x="default payment next month", y="PAY_0", data=data)
plt.title("Payment Status (PAY_0) vs Default")
plt.xlabel("Default Status")
plt.ylabel("PAY_0 (Payment Delay)")
plt.show()
```

```python
# Average payment amounts by default status
pay_cols = ["PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]
pay_means = data.groupby("default payment next month")[pay_cols].mean()
plt.figure(figsize=(12, 6))
pay_means.T.plot()
plt.title("Average Payment Amounts Over Time by Default Status")
plt.xlabel("Payment Month")
plt.ylabel("Average Payment Amount")
plt.legend(["Non-Default (0)", "Default (1)"])
plt.show()
```

### EDA Observations
- **Dataset**: 30,000 records, 25 features, no missing values.
- **Default Rate**: ~22% of customers default, indicating class imbalance.
- **Credit Limit**: Defaulters tend to have lower `LIMIT_BAL`, suggesting credit constraints.
- **Demographics**: 
  - Females (SEX=2) slightly more frequent among defaulters than males (SEX=1).
  - Age distribution similar, but younger customers (<30) show slightly higher default risk.
- **Payment History**: `PAY_0` delays strongly correlate with defaults (higher medians for 1s).
- **Payment Amounts**: Defaulters consistently pay less across months (`PAY_AMT1` to `PAY_AMT6`).
- **Correlations**: `PAY_0` to `PAY_6` and `BILL_AMT` features show clustering, indicating payment behavior is key.
